```python
import random
import time as t
%matplotlib inline
import matplotlib.pyplot as plt
q=[]
r=[]
s=[]
u=[]
arr=[]
z=0
avtimelist_b=[]
avtimelist_i=[]
avtimelist_m=[]
avtimelist_q=[]

timelist_b=[]
timelist_i=[]
timelist_m=[]
timelist_q=[]

def bubblesort(list1):
	#print(list1)

	for i in range(len(list1)):
		for j in range(len(list1)-1):
			if(list1[j]>list1[j+1]):
				list1[j],list1[j+1]=list1[j+1],list1[j]

	return (list1)


#list2=[10,25,8,11,37,6,19,28]

def insertionsort(list2):
	
	for i in range(1, len(list2)): 
  
	        key = list2[i] 
	        j = i-1
	        while j >=0 and key < list2[j] : 
	                list2[j+1] = list2[j] 
	                j -= 1
	        list2[j+1] = key
	 
	return (list2)

#print (insertionsort(list2))


def mergeSort(nlist):
    #print("Splitting ",nlist)
    if len(nlist)>1:
        mid = len(nlist)//2
        lefthalf = nlist[:mid]
        righthalf = nlist[mid:]

        mergeSort(lefthalf)
        mergeSort(righthalf)
        im=jm=km=0       
        while im < len(lefthalf) and jm < len(righthalf):
            if lefthalf[im] < righthalf[jm]:
                nlist[km]=lefthalf[im]
                im=im+1
            else:
                nlist[km]=righthalf[jm]
                jm=jm+1
            km=km+1

        while im < len(lefthalf):
            nlist[km]=lefthalf[im]
            im=im+1
            km=km+1

        while jm < len(righthalf):
            nlist[km]=righthalf[jm]
            jm=jm+1
            km=km+1 

            
def quickarrange(arr,left,right):
	pivot=right
	right=right-1
	while(left<=right):
		if(arr[left]<=arr[pivot]):
			left=left+1
		elif(arr[right]>=arr[pivot]):
			right=right-1
            
            
		else:
			arr[left],arr[right]=arr[right],arr[left]
            
	
	arr[left],arr[pivot]=arr[pivot],arr[left]
	return (left)
def quicksort(arr,left,right):

	if (right-left)>0:
		partition=quickarrange(arr,left,right)
		quicksort(arr,left,partition-1)
		quicksort(arr,partition+1,right)
        

#arr = [12, 11, 13, 5, 6, 7,85,32,45,0,-56,-12] 

#print ("\n\nSorted array is")


x=[1000,5000,10000,15000,20000,25000,30000]
#print(x)
for r in range(len(x)):

	for n in range(0,3):

		for k in range(x[z]):
			p=random.randint(0,100)
			q.append(p)
		r=q[:]
		s=q[:]
		u=q[:]

		a=t.time()
		bubblesort(q)
		b=t.time()
		
		c=t.time()
		insertionsort(r)
		d=t.time()
        
		e=t.time()
		mergeSort(s)
		f=t.time()
        
		g=t.time()
		quicksort(u,0,len(u)-1)
		h=t.time()
		
		timediff_b=b-a
		timediff_i=d-c
		timediff_m=f-e
		timediff_q=h-g
		
		timelist_b.append(timediff_b)
		timelist_i.append(timediff_i)
		timelist_m.append(timediff_m)
		timelist_q.append(timediff_q)
		del q[:]
		del r[:]
		del s[:]
		del u[:]

		#print(timelist,end="")

	avtimelist_b.append((timelist_b[0]+timelist_b[1]+timelist_b[2])/3)
	avtimelist_i.append((timelist_i[0]+timelist_i[1]+timelist_i[2])/3) 
	avtimelist_m.append((timelist_m[0]+timelist_m[1]+timelist_m[2])/3)
	avtimelist_q.append((timelist_q[0]+timelist_q[1]+timelist_q[2])/3)

	del timelist_b[:]
	del timelist_i[:]
	del timelist_m[:]
	del timelist_q[:]

	timelist_b=[]
	timelist_i=[]
	timelist_m=[]
	timelist_q=[]

	z=z+1

#print(avtimelist_m)
#plt.yscale('linear',linthreshy=0.01)
#plt.ylim(0,5)
fig = plt.figure(figsize=(18,13))
plt.subplot(121)


plt.plot(x,avtimelist_b,'r')
plt.plot(x,avtimelist_i,'b')
plt.plot(x,avtimelist_m,'g')
plt.plot(x,avtimelist_q,'m')

plt.grid()

for a1,a2 in zip(x[4:],avtimelist_b[4:]):
    label="{:.4f}".format(a2)
    plt.annotate(label,(a1,a2), textcoords="offset points", xytext=(0,10),ha='center') 
    
for a1,a2 in zip(x[4:],avtimelist_i[4:]):
    label="{:.4f}".format(a2)
    plt.annotate(label,(a1,a2), textcoords="offset points", xytext=(0,10),ha='center') 
    
for a1,a2 in zip(x,avtimelist_m):
    label="{:.4f}".format(a2)
    plt.annotate(label,(a1,a2), textcoords="offset points", xytext=(0,10),ha='center') 

plt.subplot(122)
plt.ylim(0,1.2)
plt.plot(x,avtimelist_b,'r')
plt.plot(x,avtimelist_i,'b')
plt.plot(x,avtimelist_m,'g')
plt.plot(x,avtimelist_q,'m')


plt.show()


# RED CURVE  = BUBBLE SORT
# BLUE CURVE = INSERTION SORT
# GREEN CURVE = MERGE SORT
# MAGENTA CURVE= QUICK SORT
#ANOTHER GRAPH IS TO SHOW DIFFERENCE BETWEEN QUICK AND MERGE SORT
```
